```python
import matplotlib.pyplot as plt

# Data dari grafik
x = [1, 2, 3, 4]
hw_xrar = [1.02, 0.98, 1.015, 1.005]
max_xrar = [1.18, 1.08, 1.10, 1.08]
error = [0.065, 0.065, 0.065, 0.065]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of layers')
ax1.set_ylabel('XRAR Performance', color=color)
ax1.errorbar(x, hw_xrar, yerr=[0.03]*4, fmt='-o', color=color, label='HW-XRAR')
ax1.plot(x, max_xrar, 'x--', color=color, label='HW Max-XRAR')
ax1.axhline(1.0, color='red', linestyle=':', label='HW-XRAR bound (1.0)')
ax1.tick_params(axis='y', labelcolor=color)

# Annotasi di XRAR
ax1.annotate('Initial overshoot', xy=(1, 1.02), xytext=(1.5, 1.07),
             arrowprops=dict(facecolor='red', arrowstyle="->"),
             fontsize=8, color='red')
ax1.annotate('Min XRAR', xy=(2, 0.98), xytext=(2.5, 0.95),
             arrowprops=dict(facecolor='red', arrowstyle="->"),
             fontsize=8, color='red')
ax1.annotate('Stable region', xy=(3.5, 1.01), xytext=(3, 1.08),
             arrowprops=dict(facecolor='red', arrowstyle="->"),
             fontsize=8, color='red')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Normalized Error', color=color)
ax2.plot(x, error, 'd--', color=color, label='HW Error')
ax2.axhline(0.06, color='blue', linestyle=':', label='HW Error Bound (0.06)')
ax2.tick_params(axis='y', labelcolor=color)

# Annotasi di error
ax2.annotate('Consistently low error', xy=(2, 0.065), xytext=(0.5, 0.08),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=8, color='blue')

fig.tight_layout()
fig.legend(loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.3))

plt.show()
```
