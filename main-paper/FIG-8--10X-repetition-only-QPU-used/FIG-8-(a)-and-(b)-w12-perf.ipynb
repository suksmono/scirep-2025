```python
from matplotlib import pyplot as plt
import numpy as np
###############################################################
# plot performance curves
# SIM: simulation data
# >>> SIM-XRAR by number of layer + stdev/sqrtN
# >>>>> upper bound: 4.0000
# >>> SIM-ERROR of Williamson cost function
# >>>>> lower bound: 0.0000
# HW: QPU data
# >>> HW-XRAR by number of layer
# >>>>> random algo lower-bound: 1.0000
# >>> HW-ERROR of Williamson cost function
# >>>>> random algo lower-bound: 0.0000
###############################################################


#y_eq_err= np.array([0.05556, 0.05556, 0.05556, 0.05556,0.05556])

###############################################################
## -- data xRAR SIMULASI --
###############################################################
N=10
sqrtN=np.sqrt(N)
x=np.array([1,4,8,16, 32])
sim_xrar = np.array([2.4270, 2.9621, 3.6477, 3.8984,3.9414])
sim_xrar_stdev=np.array([0.1124/sqrtN, 1.0665/sqrtN, 0.2418/sqrtN, 0.0996/sqrtN, 0.0374/sqrtN])
### upper bound of XRAR for Williamson-12
y_eq4= np.array([4, 4, 4, 4,4])


############
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

#-- plot x1
ax1.plot(x, sim_xrar, 'b-o')
ax1.errorbar(x, sim_xrar, yerr=sim_xrar_stdev, fmt='bd:')
ax1.plot(x, y_eq4, 'b-')

ax1.set_xlabel('number of layers')
ax1.set_ylabel('XRAR performance')
##
#-- plot x2
## data -error bin
sim_error = np.array([0.0217, 0.0143, 0.0078, 0.0025, 0.0022])
sim_error_stdev=np.array([0.0092/sqrtN, 0.0073/sqrtN, 0.0046/sqrtN, 0.0018/sqrtN, 0.0010/sqrtN])

# lower bound of obj-error
y_eq0= np.array([0, 0, 0, 0,0])

ax2.plot(x, sim_error, 'r:d')
ax2.errorbar(x, sim_error, yerr=sim_error_stdev,fmt ='ro:')
ax2.plot(x, y_eq0, 'r:')
ax2.set_ylabel('normed error')


#ax1.legend(["SIM-XRAR"], loc ="center")
ax1.legend(["SIM-XRAR", "SIM-XRAR-bound"], loc ="center right")
ax2.legend(["SIM-error", "SIM-error-bound"], loc ="lower right")

plt.show()

###############################################################
## -- data xRAR Hardware QPU --
###############################################################
# >>> HW-XRAR by number of layer
# >>>>> random algo lower-bound: 1.0000
# >>> HW-ERROR of Williamson cost function
# >>>>> random algo lower-bound: 0.0000
###############################################################
fig2, ax1 = plt.subplots()
ax2 = ax1.twinx()
N2=4
sqrtN2=np.sqrt(N2)
qpu_x=np.array([1,2,3,4])
qpu_xrar_mean = np.array([1.0211, 0.9863, 1.0078, 1.0012])
qpu_xrar_stdev = np.array([0.0960/sqrtN2, 0.0603/sqrtN2, 0.0464/sqrtN2, 0.0657/sqrtN2])
# max data
qpu_xrar_max = np.array([1.1836, 1.0781, 1.0977, 1.0938])

## plot mean
ax1.plot(qpu_x, qpu_xrar_mean, 'r-d')
ax1.errorbar(qpu_x, qpu_xrar_mean, yerr=qpu_xrar_stdev, fmt='rd-')
## plot max
ax1.plot(qpu_x, qpu_xrar_max, 'r-.x')
ax1.set_xticks(qpu_x)
### mean for random alg: xRAR =1
y_eq1=np.array([1,1,1,1])
ax1.plot(qpu_x, y_eq1, 'r-')
#
ax1.set_xlabel('number of layers')
ax1.set_ylabel('XRAR Performance')

###################################
### plot HW-error
##-- threshold: random algorithm
qpu_error = np.array([0.0536, 0.0547, 0.0547, 0.0543])
qpu_error_stdev = np.array([0.0012/sqrtN2, 0.0005/sqrtN2, 0.0008/sqrtN2, 0.0009/sqrtN2])
## lower bound error by random algo
##y_eq_err= np.array([0.05556, 0.05556, 0.05556, 0.05556])
y_eq_err= np.array([0.2222, 0.2222, 0.2222, 0.2222])

ax2.plot(qpu_x, qpu_error, 'b:x')
ax2.errorbar(qpu_x, qpu_error, yerr=qpu_error_stdev,fmt ='bx:')
ax2.plot(qpu_x, y_eq_err, 'b:')
## plot lower/random bound

ax2.set_ylabel('normed error')

ax1.legend(["HW-XRAR","HW Max-XRAR", "HW-XRAR-bound"], loc ="center left" )
ax2.legend(["HW-error" ,"HW-error-bound"],loc ="upper right"  )
##
plt.show()
```
