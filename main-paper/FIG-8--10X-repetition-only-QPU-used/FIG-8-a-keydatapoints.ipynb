```python
import matplotlib.pyplot as plt

# asumsi variabel x, y, error_y sudah sesuai
x = [1, 5, 8, 16, 32]
xrar = [2.4, 3.0, 3.65, 3.9, 3.92]
err = [0.2, 0.25, 0.3, 0.15, 0.05]
error = [0.028, 0.018, 0.012, 0.004, 0.004]

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Number of layers')
ax1.set_ylabel('XRAR performance', color=color)
ax1.errorbar(x, xrar, yerr=err, fmt='-o', color=color, label='XRAR performance')
ax1.axhline(4, color='blue', linestyle='--', label='XRAR upper bound (4)')
ax1.tick_params(axis='y', labelcolor=color)

# Annotasi
ax1.annotate('Start (Low XRAR)', xy=(1, 2.4), xytext=(3, 2.6),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=8, color='blue')
ax1.annotate('Significant increase', xy=(8, 3.65), xytext=(10, 3.2),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=8, color='blue')
ax1.annotate('Near optimal', xy=(16, 3.9), xytext=(18, 3.6),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=8, color='blue')
ax1.annotate('Stable performance', xy=(32, 3.92), xytext=(28, 3.75),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=8, color='blue')

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Normalized error', color=color)
ax2.plot(x, error, 'd--', color=color, label='Normalized error')
ax2.axhline(0, color='red', linestyle=':')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()

fig.legend(loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.25))

plt.show()
```
